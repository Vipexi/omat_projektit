```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
train = "sm_train.jpg"
test1 = "sm_test1.jpg"
test2 = "sm_test2.jpg"
```

```python
img_test1 = cv2.imread(test1,cv2.IMREAD_GRAYSCALE)
img_train = cv2.imread(train, cv2.IMREAD_GRAYSCALE)
```

```python
orb = cv2.ORB_create()
key_points_train, des_train = orb.detectAndCompute(img_train, None)
key_points_test1, des_test1 = orb.detectAndCompute(img_test1, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des_train, des_test1)
matches = sorted(matches, key=lambda x:x.distance)
img_final = cv2.drawMatches(img_train, key_points_train, img_test1, key_points_test1,matches[:10], None)
plt.imshow(img_final), plt.show
```
