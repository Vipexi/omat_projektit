```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img_ros = "ros.png"
img_ros_i = "ros_i.png"
```

```python
img_ros = cv2.imread(img_ros,cv2.IMREAD_GRAYSCALE)
img_ros_i = cv2.imread(img_ros_i, cv2.IMREAD_GRAYSCALE)
```

```python
orb = cv2.ORB_create()
key_points_ros, des_ros = orb.detectAndCompute(img_ros, None)
key_points_ros_i, des_ros_i = orb.detectAndCompute(img_ros_i, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des_ros, des_ros_i)
matches = sorted(matches, key=lambda x:x.distance)
img_final = cv2.drawMatches(img_ros, key_points_ros, img_ros_i, key_points_ros_i,matches[:10], None)
plt.imshow(img_final), plt.show
```
